```python
#Sets
colors = {'Green', 'Green', 'Red', 'Blue', 'Indigo', 'Violet', 'Black'}
print(colors)
```

the difference between a set, list and turple, sets are unordered sequence of items while list and tuple are ordered, a set doesn't allow a duplicate of items unlike list and tuple that allow that.

the above example of a set had Green duplicated and when creating the set the output exempted one of the duplicate item as seen above

```python
#Using a Set function
colors = set(['red', 'white', 'black', 'green'])
print(colors)
```

Using a set function as seen above can convert a list to a set 

```python
#Using Set and List functions

colors = set(['red', 'black', 'green', 'blue', 'green'])
colors = list(colors)

print(colors)
```

the idea behind the above function was to remove the duplicate item (green) from the list using the set function and converting back to a list 

```python
# Adding items to a set
colors = set(['red', 'black', 'green', 'white', 'black'])
colors.add('Blue')
print(colors)
```

so Set has the "add" method which can add an item to a set. 

```python
#Removing items from a set
colors = set(['red', 'blue', 'white', 'green', 'black'])
colors.remove('red')
print(colors)
```

Just as the name 'remove' implies, it removes an item from a set but there are 2 ways to remove items from a set which are using the method, 'remove and discard'. lets see what discard looks like.

```python
colors = set(['red', 'blue', 'white', 'green', 'black'])
colors.discard('red')
print(colors)
```

```python
#Find set size
colors = set(['red', 'black', 'green', 'blue', 'orange'])
colors_2 = {'black', 'red', 'green', 'orange'}

print(len(colors))
print(len(colors_2))
```

so the above code does the exact function its name depicts, as we already know "len" shows the length of a set, list or tuple, but above i used the 2 ways that a set can be written to create 2 sets and find how many items the both sets had.

```python
#Set Operation
#Set Union

A = {'red', 'green', 'blue'}
B = {'yellow', 'red', 'orange'}

#by operator
print(A|B)
```

So mathematically a Union will combine the elements of 2 sets but would not combine the duplicate, only one will exist in the new set and this is what happened above.

```python
A = {'red', 'green', 'blue'}
B = {'yellow', 'red', 'orange'}

#by method
print(A.union(B))
```

so there are 2 ways to create a union, one is by using the operator and the other is by using a method.

```python
#Set intersection
A = {'red', 'green', 'blue'}
B = {'yellow', 'red', 'orange'}

#by operator 

print(A & B)
```

```python
A = {'red', 'green', 'blue'}
B = {'yellow', 'red', 'orange'}

#by method

print(A.intersection (B))
```

```python
#Set Difference
A = {'red', 'green', 'blue'}
B = {'yellow', 'red', 'orange'}

#by operator
print(A - B)
```

```python
A = {'red', 'green', 'blue'}
B = {'yellow', 'red', 'orange'}

#by method
print(A.difference(B))
```

```python
#Computing Symmetry 
A = {'red', 'green', 'blue'}
B = {'yellow', 'red', 'orange'}

#by operator
print(A ^ B)
```

```python
# Python Functions

def Hello():
  print("Hello World")

#Passing an Agrument to the function
def Hello(name):
  print("Hello World")

Hello('Bob')
```

Now note carefully that the first function had no positional argument while the second had a positional argument passed and this enables one pass in Bob to test the function, but let us look at the types of arguments and how to use it shall we?

```python
#Types of Arguments
#Positional Arguments
#Keyword Arguments
#Default Arguments
#Variable Length Positional Arguments(*args)
#Variable Length Keyword (**Kwargs)
```

```python
#Positional Arguments
def func(name, job):
  print(name, 'is a', job)

func('Bob', 'developer')
```

```python
#Keyword Arguments
def func(name, job):
  print(name, 'is a', job)
func(name = 'Bob', job = 'developer')
```

the slight difference between keyword and positional arguments is just the fact that you call the argument like a variable and assign the object/value to it which is seen when calling the function above. this is to avoid confusion of positional argument.

```python
#Default argument
def func(name, job = 'developer'):
  print(name, 'is a', job)

func('Bob', 'manager')
func('Bob')
```

these little tricky things are what makes python a specical language to me, now writing the function a default argument was passed as (job ='developer'), so when calling the function if you pass the title of a job to the function it prints the value "Bob is a manager" and if you don't python return the value with the default argument passed in the function "Bob is a developer"

```python
#Variable Length Arguments(*args and **KWargs)
def unlimited_arguments(*args):
  print(args)

unlimited_arguments(1, 12, 34, 56, 87, 92)
```

Although i will most likely try to use this code in a project because i havn't fully understood how this works, but the args argument takes unlimited number of arguments, reason why i named the function unlimited arguments. but the asterick (*) all the unmatched positional arguments into a tuple and any operation can be performed on it but we will look at this in a real project 

```python
# Variable Lenght Arguments(**Kwargs)
def unlimited_arguments(**kwargs):
  print(kwargs)

unlimited_arguments(name = 'Bob', age = 25, job= 'devops' )
```

Now lets just look at the difference between args and kwargs, there is really an intuitive aspect of the above code, using kwargs, kwargs takes the keyword arguments. in the sense that you can clearly see that when the function was called the key and values where placed in to create the dictionary.
